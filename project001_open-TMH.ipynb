# Monty Hall Problem Game

## This game is going to prove the Monty Hall Problem in a funny way

### Main concepts

first: show the three options and randomly and secreatly choose an option to put the prize

second: set the prize in a door

third: get the user's option as an input

forth: open an empty door

fifth: ask if the user wants to change doors

sixth: open user's door and save results

seventh: ask if the player wants to play again

```python
import numpy as np
import math
import re
import random as rand
import os
from IPython.display import clear_output
from time import sleep
```

```python
# create variables to store the results
games_played = 1
games_changed = 0
games_not_changed = 0
wins_changed = 0
wins_not_changed = 0
# Create a function that can be called again if the player wants to continue playing
def play(games_played, games_changed, games_not_changed, wins_changed, wins_not_changed):
    '''
    This is the function that contains all the game
    '''
    clear_output(wait=True)
    # FIRST STEP: Initial interface
    initial_interface = '''
         _____     _____     _____
        |     |   |     |   |     |
        |  1  |   |  2  |   |  3  |
        |_____|   |_____|   |_____|
    '''
    print(initial_interface)

    # SECOND STEP: Set the prize in a door
    prize_door = rand.randint(1,3)
    #print(f'prize door: {prize_door}')
    # THIRD STEP: Get the user's option as an input
    print('Choose a door (1/2/3): ')
    player_door = input('Choose a door: ')
    while player_door != '1' and player_door != '2' and player_door != '3':
        player_door = input('Choose a door (1/2/3): ')
    player_door = int(player_door)
    clear_output(wait=True)
    #print(f'Player option: {player_door}')
    # FORTH STEP: Open an empty door
    empty_door = rand.randint(1,3)
    while (empty_door == prize_door or empty_door == player_door):
        empty_door = rand.randint(1,3)
    #print(f'empty door: {empty_door}')
    new_interface = re.sub(str(player_door), 'P', initial_interface)
    new_interface = re.sub(str(empty_door), ' ', new_interface)
    print(new_interface)
    print(f'You\'ve selected door {player_door} and the door {empty_door} is empty')
    # FIFTH STEP: Ask if player wants to change doors
    other_door = rand.randint(1,3)
    while (other_door == empty_door or other_door == player_door):
        other_door = rand.randint(1,3)
    #print(other_door)
    print(f'Do you want to change your option to door {other_door} (y/n)?')
    change_door = input(f'Do you want to change your option to door {other_door} (y/n)?')
    while change_door != 'y' and change_door != 'n':
        change_door = input(f'Do you want to change your option to door {other_door} (y/n)?')
    print(f'Player option: {change_door}')
    # SIXTH STEP: Show results
    new_interface = re.sub(str(prize_door), '*', initial_interface)
    new_interface = re.sub(str(empty_door), ' ', new_interface)
    new_interface = re.sub(str(other_door), ' ', new_interface)
    new_interface = re.sub(str(player_door), ' ', new_interface)
    print(new_interface)
    if change_door == 'y':
        staff_door = player_door
        player_door = other_door
        other_door = staff_door
        games_changed += 1
        if player_door == prize_door:
            wins_changed += 1
            print('Congratz! You won the prize!')
        else:
            print('Sorry! You\'ve lost!')
    else:
        games_not_changed += 1
        if player_door == prize_door:
            wins_not_changed += 1
            print('Congratz! You won the prize!')
        else:
            print('Sorry! You\'ve lost!')
    # print results
    win_percentage_changed = 0
    win_percentage_not_changed = 0
    try:
        win_percentage_changed = wins_changed/games_changed
    except ZeroDivisionError:
        win_percentage_changed = 0
    try:
        win_percentage_not_changed = wins_not_changed/games_not_changed
    except ZeroDivisionError:
        win_percentage_not_changed = 0
    print(f'''
    Total games played: {games_played}
    Games that changed doors: {games_changed}
    win % when changed: {win_percentage_changed}
    Games that player did not change doors: {games_not_changed}
    win % when did not changed: {win_percentage_not_changed}
    ''')
    sleep(1)
    # SEVENTH STEP: Ask if player wants to play again
    games_played += 1
    play_again_yn = 'x'
    while play_again_yn != 'y' and play_again_yn != 'n':
        # Ask if the user wants to play again
        play_again_yn = input('Play again (y/n)? ')
    if play_again_yn == 'y':
        play(games_played, games_changed, games_not_changed, wins_changed, wins_not_changed)
    elif play_again_yn == 'n':
        print(f'''
        Thanks for playing!
        ''')
    
play(games_played, games_changed, games_not_changed, wins_changed, wins_not_changed)
#print(games_played)
```
